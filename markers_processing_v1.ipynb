```python
def plot_markers(data):
    data.drop_sel(axis = "ones").plot.line(x = "time", col = "channel", hue = "axis", col_wrap = 3)
```

```python
plot_markers(readed_markers)
plot_markers(readed_markers2)
```

```python
def plot_one_marker(data, axis, channel_num):
        data.sel(axis = axis, channel = data['channel'][channel_num]).plot.line(x="time")
```

```python
plot_one_marker(readed_markers, "y", 36)
plot_one_marker(readed_markers2, "y", 36)
```
