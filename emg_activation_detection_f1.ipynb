```python
muscles_names = [
    "Biodrowo-żebrowy - wyżej L",
    "Biodrowo-żebrowy - niżej L", 
    "Skośny brzucha L", 
    "Prosty uda L",
    "Dwugłowy uda L", 
    "Brzuchaty łydki - głowa boczna L",
    "Brzuchaty łydki - głowa przyśrodkowa L", 
    "Płaszczowy L",
    "Biodrowo-żebrowy - wyżej P",
    "Biodrowo-żebrowy - niżej P", 
    "Skośny brzucha P", 
    "Prosty uda P",
    "Dwugłowy uda P", 
    "Brzuchaty łydki - głowa boczna P",
    "Brzuchaty łydki - głowa przyśrodkowa P", 
    "Płaszczowy P",]

def read_analogs(data_path):
    muscles = [
    "Voltage.1",
    "Voltage.2",
    "Voltage.3",
    "Voltage.4",
    "Voltage.5",
    "Voltage.6",
    "Voltage.7",
    "Voltage.8",
    "Voltage.9",
    "Voltage.10",
    "Voltage.11",
    "Voltage.12",
    "Voltage.13",
    "Voltage.14",
    "Voltage.15",
    "Voltage.16"]
    
    emg = Analogs.from_c3d(data_path, usecols=muscles)
    
    emg['channel']=muscles_names
    return emg
```

```python
def normalize_data(data_path):
    emg_p = (
        data_path.meca.band_pass(order=2, cutoff=[10, 425])
        .meca.center()
        .meca.abs()
        .meca.low_pass(order=4, cutoff=5, freq=data_path.rate)
        .meca.normalize(ref=None, scale=1)
    ) 
    emg_p.name = "EMG"
    emg_p.attrs["units"] = "%"
    emg_p.time.attrs["units"] = "seconds"
    return emg_p
```

```python
def read_events(data_path):
    c3d_to_compare= c3d(data_path)
    event_time = c3d_to_compare['parameters']['EVENT']['TIMES']['value'][1]
    
    return event_time
```

```python
def activation_detection(data, rounded_events, normalized_emg):
    ii = 0
    print(data)
    while(ii < 16):
        start_event = 0
        end_event = 1
        j = 0
        print("\n")
        print(muscles_names[ii])
        print("\n")
        while(j < 7):
            i = rounded_events[start_event]
            check = 0
            while(i < rounded_events[end_event]):
                if normalized_emg[ii][i] > 0.25:
                    #print("yes")
                    check = check+1
                    if check == 3:
                        if start_event == 0:
                            print("Mięsień aktywny podczas - Wyciągnięcia do pacjenta")
                        if start_event == 2:
                            print("Mięsień aktywny podczas - Ugięcie nóg pacjenta")
                        if start_event == 4:
                            print("Mięsień aktywny podczas - Ułożenie prawej ręki pacjenta")
                        if start_event == 6:
                            print("Mięsień aktywny podczas - Ułożenie lewej ręki pacjenta")
                        if start_event == 8:
                            print("Mięsień aktywny podczas - Przewrócenie pacjenta")
                        if start_event == 10:
                            print("Mięsień aktywny podczas - Opuszczenie nóg pacjenta")
                        if start_event == 12:
                            print("Mięsień aktywny podczas - Posadzenie pacjenta")
                        break
                i = i+100
            j = j+1
            start_event = start_event+2
            end_event = end_event+2
        ii = ii+1
```
